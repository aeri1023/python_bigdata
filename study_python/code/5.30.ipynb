```python
html = urllib.request.urlopen('http://www.cgv.co.kr/movies/?ft=0')
```

```python
soup=BeautifulSoup(html,"lxml")
```

```python
ol=soup.find("ol")
li=ol.find_all("li")
for l in li:
    div = l.find("div",{"class":"box-contents"})
    strong =div.find("strong").text
    print(strong)
```

```python
import urllib
from bs4 import BeautifulSoup
```

```python
output='output.txt'
open_output_file =open(output,'w')

URL = 'http://news.naver.com/main/read.nhn?mode=LSD&mid=shm&sid1=105&oid=001&aid=0010116602'
source_code_from_URL=urllib.request.urlopen(URL)
soup=BeautifulSoup(source_code_from_URL,"lxml",from_encoding='utf-8')

text=''

for item in soup.find_all('div', id = 'articleBodyContents'):
    
    text= text+str(item.find_all(text=True))
    
    open_output_file.write(text)
    open_output_file.close()
```

```python
import re
input_file='output.txt'
output_file='output_cleaned.txt'
read_file = open(input_file,'r')
write_file = open(output_file,'w')

text= read_file.read()
print(text)

cleaned_text=re.sub('[a-zA-Z]','',text)
print(cleaned_text)
```

```python
cleaned_text=re.sub('[\{\}\[\]\/?.,;:|\)*~`!^-_+<>@\#$%&\\\=\(\'\"]','',cleaned_text)
```

```python
write_file.write(cleaned_text)
read_file.close()
write_file.close()
```
