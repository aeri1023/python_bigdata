```python
station_name=[] #빈리스트 생성
for name in crime_police['관서명']:
    station_name.append('서울'+str(name[:-1])+'경찰서') #이름들을 하나씩 가저와 name에 넣고 맨뒷글자를 슬라이스 하여 문자열 지정, 서울  경찰서로 배열에 추가됨 
    
station_name
```

```python
station_add=[] #주소리스트 생성
station_lat=[] #위도리스트
station_lng=[] #경도리스트
for name in station_name: #station_name 리스트를 하나씩 가져와 name변수에 넣고
    tmp=gmaps.geocode(name, language='ko') # 경찰서 이름으로 구글맵에서 검색한 정보를 tmp변수에 넣음
    station_add.append(tmp[0].get("formatted_address")) #station_add에 formatted_address정보를 넣어줌
    
    tmp_loc=tmp[0].get("geometry")
    station_lat.append(tmp_loc['location']['lat']) #lat과 lng에 loc의 location과 lat, lng 넣음
    station_lng.append(tmp_loc['location']['lng'])
    
    print(name+'---->'+tmp[0].get("formatted_address")) #주소 형태로 프린트
    
```

```python
#----------------------------------------5.16일 내용 이어서 앞에 내용 실행시키기
```

```python
gu_name=[]
for name in station_add: #station_add의 수만큼 돌림
    tmp=name.split() #split 띄어쓰기와 공백 제외
    tmp_gu=[gu for gu in tmp if gu[-1]=='구'][0] #마지막 글자가 구인 부분을 리스트화 그 리스트의 0번째를 tmp_gu에 넣어줌(어차피 구는 한단어밖에 안나올 것이므로)
    gu_name.append(tmp_gu)

crime_police['구별']=gu_name  #crime_anal_police를 나는 crime_police로 저장함
crime_police.head()
```
