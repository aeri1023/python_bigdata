```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

def sales_clean():
    sales = pd.read_csv(u'서울시 동물판매업 현황.csv', encoding='utf-8')
    guname=[]
    size = int(sales[u'도로명전체주소'].shape[0])
    for name in range (0,size):
            try:
                if (sales.iloc[name][u'도로명전체주소']==None)==True:
                    gu= sales.iloc[name][u'도로명전체주소'].split(' ')[1]
                    guname.append(gu)
                else:
                    gu= sales.iloc[name][u'소재지전체주소'].split(' ')[1]
                    guname.append(gu)
            except : 
                guname.append(gu)
    gn=pd.DataFrame({'gu':guname})
    sales = sales.join(gn,how='outer')
    sales=sales.loc[(sales[u'영업상태명'].str.contains(u'운영중'))][[u'사업장명',u'소재지전체주소',u'도로명전체주소',u'영업상태명',u'gu']]
    sal=list(sales['gu'].dropna())
    return sal

def fre_sales(list_sales):
    d=dict()
    for sent in list_sales:
        for w in sent.split():
            if w not in d:
                d[w]=1
            else:
                d[w]=d[w]+1
    d_value = d.values()
    d_value.sort(reverse=True)
    num = d_value[10]
    d1 = dict()
    for key, value in d.iteritems():
        if value>=num:
            d1[key]=value
    plt.xlabel(u'동물판매업',fontsize=14)
    plt.ylabel(u'구별',fontsize=14)
    plt.barh(range(len(d1.values())), d1.values(), color = 'darkblue')
    plt.yticks(range(len(d1.values())),d1.keys())
    plt.savefig("sales_gu.png")
    plt.show()

def main():
    fre_sales(sales_clean())
if __name__=="__main__":
    main()
```
