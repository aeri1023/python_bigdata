```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

def medical_clean():
    medical = pd.read_csv(u'서울시 동물용 의료용구 판매업 현황.csv',encoding='utf-8')
    guname=[]
    size = int(medical[u'도로명전체주소'].shape[0])
    for name in range (0,size):
            try:
                if (medical.iloc[name][u'도로명전체주소']==None)==True:
                    gu= medical.iloc[name][u'도로명전체주소'].split(' ')[1]
                    guname.append(gu)
                else:
                    gu= medical.iloc[name][u'소재지전체주소'].split(' ')[1]
                    guname.append(gu)
            except : 
                guname.append(None)
    gn=pd.DataFrame({u'gu':guname})
    medical = medical.join(gn,how='outer')
    medical=medical.loc[(medical[u'영업상태명'].str.contains(u'운영중'))][[u'사업장명',u'소재지전체주소',u'도로명전체주소',u'영업상태명',u'gu']]
    med=list(medical[u'gu'].dropna())
    return med

def fre_medical(list_medical):
    d=dict()
    for sent in list_medical:
        for w in sent.split():
            if w not in d:
                d[w]=1
            else:
                d[w]=d[w]+1
    d_value = d.values()
    d_value.sort(reverse=True)
    num = d_value[10]
    d1 = dict()
    for key, value in d.iteritems():
        if value>num:
            d1[key]=value
    plt.xlabel(u'의료용구',fontsize=14)
    plt.ylabel(u'구별',fontsize=14)
    plt.barh(range(len(d1.values())), d1.values(), color = 'darkblue')
    plt.yticks(range(len(d1.values())),d1.keys())
    plt.savefig("medical_gu.png")
    plt.show()
    
def main():
    fre_medical(medical_clean())
if __name__=="__main__":
    main()
```
