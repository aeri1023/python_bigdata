```python
import time
import pandas as pd
import os
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup as bs
import random
def crawl_you(url,file_name):#url은 리스트 형태로
    ar_list=[]
    for u in url:
        pageload(driver,u,ar_list)
        print(u)
    df = pd.DataFrame(ar_list)
    df.to_csv(file_name, index=False)
def pageload(driver,url,ar_list):
    driver.get(url)
    body = driver.find_element_by_tag_name("body")
    print('시작')
    #driver.implicitly_wait(10)
    #driver.find_element_by_css_selector('#movie_player > div.ytp-chrome-bottom > div.ytp-chrome-controls > div.ytp-left-controls > button').click()
    num_of_pagedowns = 50
    while num_of_pagedowns:
        body.send_keys(Keys.PAGE_DOWN)
        time.sleep(random.randint(1,2))
        num_of_pagedowns -= 1 
        try:
            driver.find_element_by_xpath('//*[@id="more"]/span').click()
        except:
            continue
    crawl(driver,ar_list)

def crawl(driver,ar_list):

    soup = bs(driver.page_source, 'html.parser')
    article_list = driver.find_elements_by_css_selector('#content-text')
    #title = driver.find_elements_by_css_selector('#container > h1 > yt-formatted-string')
    #title = title[0].text
    for i in article_list:
        content = i.text.strip()
    #    ar_list.append({'title':title, 'content':content}) 
        ar_list.append({'content':content})
    
    #columnList=['title','content']
    #df =df[columnList]
    return ar_list
```

```python
def make_wordcount_you(file_in, file_out):
    import numpy as np
    import pandas as pd
    from pandas import DataFrame
    from konlpy.tag import Twitter
    from collections import Counter

    man = pd.read_csv(file_in,sep=',',encoding='utf-8')
    content=''
    for c in man['content']:
        content+=c.strip()
    print("전체 수:",len(man))
    import re
    hangul = re.compile('[^ \u3131-\u3163\uac00-\ud7a3]+')
    content= hangul.sub(' ', content)
    spliter=Twitter()
    nouns = spliter.nouns(content)
    count = Counter(nouns)
    return_list = []
    for n,c in count.most_common(len(count)):
        if len(n) != 1:
            temp = {'word': n, 'count': c}
            return_list.append(temp)
    word_count = pd.DataFrame(return_list)
    columnList=['word','count']
    word_count =word_count[columnList]
    word_count.to_csv(file_out, index=False,encoding='cp949')
    print('ok')
```

```python
def make_worddf_you(file_in, file_out):
    import numpy as np
    import pandas as pd
    from pandas import DataFrame
    from konlpy.tag import Twitter
    from collections import Counter
    import re

    man = pd.read_csv(file_in,sep=',',encoding='utf-8')
    set=[]
    for c in man['content']:
        hangul = re.compile('[^ \u3131-\u3163\uac00-\ud7a3]+')
        c= hangul.sub(' ', c)
        spliter_=Twitter()
        nouns_ = spliter_.nouns(c)
        set.append(nouns_)
    print("전체 수:",len(man))
    word_set=[]
    for set1 in range(0,len(set)-1) :
        for set2 in range(1,len(set)-2):
            temp=''
            for i in set[set1]:
                if i in set[set2]:
                    if i not in temp:
                        if temp=='':
                            temp+=i
                        else :
                            temp=temp+'+'+i
            if '+' in temp:
                    word_set.append({'word':temp,'count':1})
    word_df = pd.DataFrame(word_set)
    word_df = pd.DataFrame(word_df['word'].value_counts())
    word_df.to_csv(file_out, index=True,encoding='cp949')
    print('ok')
    
```
