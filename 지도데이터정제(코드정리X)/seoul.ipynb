```python
import pandas as pd
```

```python
bus = pd.read_csv("bus.csv",encoding='utf-8')
```

```python
bus.head(3)
```

```python
for i in range(1,24):
    if i>=10:
        del bus['AG_'+str(i)+'TM']
        del bus['PS_'+str(i)+'TM']
    elif i<7:
        del bus['AG_0'+str(i)+'TM']
        del bus['PS_0'+str(i)+'TM']
del bus['BB_RT']
del bus['ROUTE_NM']
del bus['ROUTE_NO']
del bus['STTN_ARSNO']
del bus['SubPolygonID']
del bus['TGRP_CODE']
del bus['AG_00TM']
```

```python
bus.head()
```

```python
df=pd.DataFrame(bus)
```

```python
busname = df.groupby("STN_NM")
mean = busname["AG_07TM","AG_08TM","AG_09TM","PS_07TM","PS_08TM","PS_09TM"].mean()
lotate = busname["Latitude","Longitude"].median()
count = busname["PointID","ID"].count()
```

```python
busgroup=pd.merge(pd.merge(lotate,count, on="STN_NM"),mean, on="STN_NM")
del busgroup["ID"]
```

```python
busgroup.head()
```

```python
busgroup.rename(columns={"PointID":"countbus"}, inplace=True)
```

```python
ag_per=pd.DataFrame((busgroup[["AG_07TM","AG_08TM","AG_09TM"]].sum(axis=1)/busgroup["countbus"])/3)
ag_per.rename(columns={0:"ag_per"}, inplace=True)
```

```python
ps_per=pd.DataFrame((busgroup[["PS_07TM","PS_08TM","PS_09TM"]].sum(axis=1)/busgroup["countbus"])/3)
ps_per.rename(columns={0:"ps_per"}, inplace=True)
```

```python
busgroup_fin=pd.merge(pd.merge(ag_per,ps_per, on="STN_NM"),busgroup[["Latitude","Longitude"]], on="STN_NM")
```

```python
busgroup_fin.head()
```

```python
busgroup_fin.to_csv("busgroup.csv",sep=',',encoding='cp949')
```
