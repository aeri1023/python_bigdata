```python
import googlemaps
import pandas as pd
import numpy as np
```

```python
gmaps_key=""
gmaps=googlemaps.Client(key=gmaps_key)
```

```python
home=pd.read_csv('home.csv',encoding='utf-8')
home.rename(columns={'서울특별시 동대문구 약령시로 107 조수경':'address'},inplace=True)
home.head()
```

```python
home1=home[0:3000]
#home1.drop()
home_ad=[]
import re
for name in home1['address']:
    name=name.split('(')[0]
    if (name[-4]!=' '):
        name = re.sub('\t', '', name)
        home_ad.append(name)
home_ad
```

```python
home_lat=[]
home_lng=[]
count=0
error=0
for name in home_ad:
    if(count%50==0):print(count)
    try:
        count+=1
        tmp=gmaps.geocode(name, language='ko')
        tmp_loc=tmp[0].get("geometry")
        home_lat.append(tmp_loc['location']['lat'])
        home_lng.append(tmp_loc['location']['lng'])
    except:
        error+=1;
```

```python
error
```

```python
home2=pd.DataFrame({'lat':home_lat, 'lng':home_lng})
home2.to_csv("home_loc.csv",sep=',',encoding='utf-8')
    
#cafe.to_csv("cafe_loc.csv",sep=',',encoding='utf-8') #csv파일로 저장(C:\Users\go113\Documents에 저장확인)
```
