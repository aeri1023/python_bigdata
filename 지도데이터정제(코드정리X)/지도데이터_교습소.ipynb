```python
import googlemaps
import pandas as pd
import numpy as np
import re
```

```python
gmaps_key="gmaps=googlemaps.Client(key=gmaps_key)
```

```python
academy=pd.read_csv('clean_data.csv',encoding='cp949')
```

```python
academy.head()
```

```python
aca_lat=[]
aca_lng=[]
count=0
for name in academy['address']:
        if(count%50==0):print(count)
        count+=1
        try:
            tmp=gmaps.geocode(name, language='ko')
            tmp_loc=tmp[0].get("geometry")
            aca_lat.append(tmp_loc['location']['lat'])
            aca_lng.append(tmp_loc['location']['lng'])
    
        except :
            aca_lat.append("na")
            aca_lng.append("na")
print("ok")
```

```python
academy.head(count)
```

```python
aca2=pd.DataFrame({'lat':aca_lat, 'lng':aca_lng})
aca3=academy.join(aca2,how='left')
```
