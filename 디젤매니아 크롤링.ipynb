```python
def cafe_crawl(keyword, filename):
    import random
    import sys
    import csv
    import codecs
    import time
    import pandas as pd
    import os
    from selenium import webdriver
    from bs4 import BeautifulSoup as bs
    base_url = "https://m.cafe.naver.com/ArticleSearchList.nhn?search.query=" + keyword + "&search.menuid=&search.searchBy=0&search.sortBy=date&search.clubid=11262350&search.option=0&search.defaultValue=1"
    driver.get(base_url)
    for page_num in range(1,50):
        driver.find_element_by_xpath('//*[@id="moreButton"]').click()
        time.sleep(1)
        driver.execute_script("nclk(this, 'cfs*n.more', '', '')")

    article_list = driver.find_elements_by_css_selector('#articleList > ul> li > a')
    article_urls = [i.get_attribute('href') for i in article_list]
    ar_list=[]

    for article in article_urls:
        try:
            driver.implicitly_wait(random.randint(3,5))
            driver.get(article)
            soup = bs(driver.page_source, 'html.parser')
            title = soup.select('h2.tit')[0].get_text()
            date = soup.select("span.date.font_l")[0].get_text()
            content_tags = soup.select('#postContent')[0].select('p')
            content = ' '.join([ tags.get_text() for tags in content_tags ])
            ar_list.append({'title' : title, 'date' : date, 'content' : content})
        
        except:
            pass
    
    cafe_df = pd.DataFrame(ar_list)
    cafe_df.to_csv(filename, index=False)
    print('ok')
```

```python
def make_wordcount(file_name_in, file_name_out):
    import numpy as np
    import pandas as pd
    from pandas import DataFrame
    from konlpy.tag import Twitter
    from collections import Counter
    import re
    name = ['content','date','title']
    man = pd.read_csv(file_name_in,sep=',',header=None,names=name)
    content_man=man['content']
    title_man = man['title']
    content_list=content_man+title_man
    content=''
    for c in content_list:
        c=str(c)
        content+=c.strip()
    #print(man.info())
    print("전체 수:",len(man))
    hangul = re.compile('[^ \u3131-\u3163\uac00-\ud7a3]+')
    content= hangul.sub(' ', content)
    spliter=Twitter()
    nouns = spliter.nouns(content)
    count = Counter(nouns)
    return_list = []
    for n,c in count.most_common(len(count)):
        if len(n) != 1:
            temp = {'word': n, 'count': c}
            return_list.append(temp)
    word_count = pd.DataFrame(return_list)
    columnList=['word','count']
    word_count =word_count[columnList]
    word_count.to_csv(file_name_out, index=False,encoding='cp949')
    print('ok')
```

```python
def make_worddf(file_name_in, file_name_out, keyword):
    import numpy as np
    import pandas as pd
    from pandas import DataFrame
    from konlpy.tag import Twitter
    from collections import Counter
    import re
    name = ['content','date','title']
    man = pd.read_csv(file_name_in,sep=',',header=None,names=name)
    set=[]
    content_man=man['content']
    title_man = man['title']
    content = title_man+content_man
    for c in content:
        c=str(c)
        hangul = re.compile('[^ \u3131-\u3163\uac00-\ud7a3]+')
        c= hangul.sub(' ', c)
        spliter_=Twitter()
        nouns_ = spliter_.nouns(c)
        set.append(nouns_)
    print("전체 수:",len(man))
    word_set=[]
    for set1 in range(0,len(set)-1) :
        for set2 in range(1,len(set)-2):
            temp=''
            for i in set[set1]:
                if i in set[set2] and  i!=keyword:
                    if i not in temp:
                        if temp=='':
                            temp+=i
                        else :
                            temp=temp+'+'+i
            if '+' in temp:
                word_set.append({'word':temp,'count':1})

    word_df = pd.DataFrame(word_set)
    word_df = pd.DataFrame(word_df['word'].value_counts())
    word_df.to_csv(file_name_out, index=True,encoding='cp949')
    print('ok')
```
